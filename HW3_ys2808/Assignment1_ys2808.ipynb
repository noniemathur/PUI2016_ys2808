```python
from __future__ import print_function
import pylab as pl
import numpy as np
from scipy.optimize import curve_fit, minimize
    
%pylab inline
```

# ASSIGNMENT 1:

### Chi Square Distribution

```python
md = {}
mymean=100
md['chisq'] = np.random.chisquare(mymean, size=100)
pl.hist(md['chisq'], bins = 30,color = 'lightblue')
pl.ylabel('N',fontsize=15)
pl.xlabel('x',fontsize=15)
pl.title('Chi Square Distribution',fontsize=15)
```

```python
print ("Chisq mean: %.2f, standard deviation: %.2f"%(md['chisq'].mean(), md['chisq'].std()))
```

```python
mysize = (2000 / (np.array(range(1, 100)))).astype(int)
```

```python
mysize = np.linspace(10,2000,100)
```

```python
md['chisq'] = {} 
for n in mysize:
    md['chisq'][n] = np.random.chisquare(mymean, size = n)
```

```python
md['chisq']['means'] = {}
axchisq_mu_n = pl.figure(figsize=(8,6)).add_subplot(111)

for nn in md['chisq']:
    if not type(nn) == str:
        md['chisq']['means'][nn] = md['chisq'][nn].mean()
        #and plot it
        axchisq_mu_n.plot(nn, md['chisq']['means'][nn], 'o')
axchisq_mu_n.set_xlabel('sample size', fontsize=18)
axchisq_mu_n.set_ylabel('sample mean', fontsize=18)
axchisq_mu_n.set_title('Chi squared', fontsize=18)
axchisq_mu_n.plot([min(mysize), max(mysize)], [mymean, mymean], 'black',linewidth = 2)
```

```python
allmeans = list(md['chisq']['means'].values())
```

```python
a=pl.figure(figsize=(8, 6)).add_subplot(111)
a.hist(allmeans,bins=30,color='lightblue')   
a.set_xlabel('sample mean', fontsize = 18)
a.set_ylabel('N', fontsize = 18)
a.set_title('Chi squared means', fontsize=18)
```

### Normal Distribution

```python
md['normal']= np.random.normal(loc=mymean,scale=1.0,size=100)
pl.hist(md['normal'], bins = 30,color = 'lightblue')
pl.ylabel('N',fontsize=15)
pl.xlabel('x',fontsize=15)
pl.title('Normal Distribution',fontsize=15)
```

```python
print ("Normal mean: %.2f, standard deviation: %.2f"%(md['normal'].mean(), md['normal'].std()))
```

```python
md['normal'] = {} 
for n in mysize:
    md['normal'][n] = np.random.normal(mymean,10,n)
md['normal']['means'] = {}
for n in mysize:
    md['normal']['means'][n] = md['normal'][n].mean()
```

```python
axnormal_mu_n = pl.figure(figsize=(8,6)).add_subplot(111)
for nn in md['normal']:
    if type(nn) != str :
        axnormal_mu_n.plot(nn, md['normal']['means'][nn], 'o')
axnormal_mu_n.set_xlabel('sample size', fontsize=18)
axnormal_mu_n.set_ylabel('sample mean', fontsize=18)
axnormal_mu_n.set_title('Normal', fontsize=18)
axnormal_mu_n.set_ylim(95,105)
axnormal_mu_n.plot([min(mysize), max(mysize)], [mymean,mymean], 'k',linewidth =2)
```

```python
allmeans = md['normal']['means'].values()
b=pl.figure(figsize=(8, 6)).add_subplot(111)
b.hist(allmeans,bins=30,color='lightblue')
b.set_xlabel('sample mean', fontsize = 18)
b.set_ylabel('N', fontsize = 18)
b.set_xlim(95,105)
b.set_title('Normal means', fontsize=18)
```

### Poisson distribution

```python
md['poisson'] = np.random.poisson(mymean, 2000)
pl.hist(md['poisson'], bins = 30,color = 'lightblue')

pl.ylabel('N',fontsize=15)
pl.xlabel('x',fontsize=15)
pl.title('Poisson Distribution',fontsize=15)
```

```python
print ("Poisson mean: %.2f, standard deviation: %.2f"%(md['poisson'].mean(), md['poisson'].std()))
```

```python
md['poisson']={}
md['poisson']['means']={}
for n in mysize:
    md['poisson'][n]= np.random.poisson(mymean, n)
    md['poisson']['means'][n]= md['poisson'][n].mean()
```

```python
plot_poi=pl.figure(figsize=(8,6)).add_subplot(111)
for nn in md['poisson']:
    if type(nn) != str:
        plot_poi.plot(nn,md['poisson']['means'][nn],'o')
plot_poi.set_xlabel('sample size', fontsize=18)
plot_poi.set_ylabel('sample mean', fontsize=18)
plot_poi.set_title('Poisson',fontsize=18)
plot_poi.plot([min(mysize),max(mysize)],[mymean,mymean],'k',linewidth=2)
```

```python
a=pl.figure(figsize=(8, 6)).add_subplot(111)
a.hist(md['poisson']['means'].values(),bins=30,color='lightblue')
a.set_xlabel('sample mean', fontsize = 18)
a.set_ylabel('N', fontsize = 18)
a.set_title('Poisson means', fontsize=18)
```

### Binomial Distribution

```python
n,p = 200,0.5
s = np.random.binomial(n, p, 100)
pl.hist(s,color='lightblue',bins=30)
pl.ylabel('N',fontsize=15)
pl.xlabel('x',fontsize=15)
pl.title('Binomial Distribution',fontsize=15)
```

```python
md['binomial']={}
md['binomial']['means']={}
plot_bi=pl.figure(figsize=(8,6),).add_subplot(111)
for i in mysize:
    md['binomial'][i]= np.random.binomial(n, p, i)
    md['binomial']['means'][i]=md['binomial'][i].mean()
    plot_bi.plot(i,md['binomial']['means'][i],'o')
plot_bi.plot([min(mysize),max(mysize)],[mymean,mymean],'k',linewidth=2)
plot_bi.set_xlabel('sample size', fontsize=18)
plot_bi.set_ylabel('sample mean', fontsize=18)
plot_bi.set_title('Binomial',fontsize=18)
```

```python
a=pl.figure(figsize=(8,6)).add_subplot(111)
a.hist(md['binomial']['means'].values(),bins=30,color='lightblue')
a.set_xlabel('sample mean', fontsize = 18)
a.set_ylabel('N', fontsize = 18)
a.set_title('Binomial means', fontsize=18)
```

### Logistic Distribution

```python
s=np.random.logistic(mymean,1.0,100)
pl.hist(s,bins=30,color='lightblue')
pl.ylabel('N',fontsize=15)
pl.xlabel('x',fontsize=15)
pl.title('Logistic Distribution',fontsize=15)
```

```python
print ("Logistic mean: %.2f, standard deviation: %.2f"%(s.mean(), s.std()))
```

```python
md['logistic']={}
md['logistic']['means']={}
plot_lo=pl.figure(figsize=(10,6)).add_subplot(111)

for n in mysize:
    md['logistic'][n]= np.random.logistic(mymean, 1.0,n)
    md['logistic']['means'][n]= md['logistic'][n].mean()
    plot_lo.plot(n,md['logistic']['means'][n],'o')
plot_lo.plot([min(mysize),max(mysize)],[mymean,mymean],'k',linewidth=2)
plot_lo.set_xlabel('sample size', fontsize=18)
plot_lo.set_ylim(99,101)
plot_lo.set_ylabel('sample mean', fontsize=18)
plot_lo.set_title('Logistic',fontsize=18)
```

```python
a=pl.figure(figsize=(8,6)).add_subplot(111)
a.hist(md['logistic']['means'].values(),bins=30,color='lightblue')
a.set_xlabel('sample mean', fontsize = 18)
a.set_ylabel('N', fontsize = 18)
a.set_xlim(99.5,100.5)
a.set_title('Logistic means', fontsize=18)
```
