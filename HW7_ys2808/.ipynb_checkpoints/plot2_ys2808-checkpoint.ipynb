```python
import numpy as np
import pandas as pd
import pylab as pl
import seaborn as sns
%pylab inline
```

```python
data=pd.read_csv('Water_Quality_complaints.csv')
data.head()
```

```python
data.rename(columns={'Incident Zip':'zipcode'},inplace=True)
```

```python
data.zipcode.dropna(inplace=True)
```

```python
data.head()
```

```python
a=data['zipcode'].groupby(data['zipcode']).count()
a
```

```python
a.sort(ascending=False)
a=pd.DataFrame(a)
a.rename(columns={'zipcode':'count'},inplace=True)
```

```python
a.iloc[0:20].plot(kind='bar')
pl.xlabel('Incident Zipcode',fontsize=15)
pl.xticks(fontsize=10)
pl.ylabel('Complaints counts',fontsize=15)
pl.title('NYC water complaints counts by zipcode,Top 20 complained',fontsize=15)
```
